```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4,5,6,7,8], dtype=float)
y = np.array([2, 3, 5, 4,5,4,7,6], dtype=float)

n = len(X)
```

```python
plt.scatter(X, y, color='blue')
plt.grid()
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data Points')
plt.show()
```

```python
# ======================
# PARÁMETROS INICIALES
# ======================
w = 0.0
b = 0.0
lr = 0.0001
```

```python
# ======================
#  GRADIENTE DESCENDENTE
# ======================
epochs =500

for i in range(epochs):
    
    # Predicción
    y_pred = w * X + b
    
    # Error
    error = y - y_pred
    
    # MSE
    mse = np.mean(error ** 2)
    
    # Gradientes
    dw = (-2/n) * np.sum(X * error)
    db = (-2/n) * np.sum(error)
    
    # Actualización
    w = w - lr * dw
    b = b - lr * db
    
    # Mostrar resultados
    print(f"Iteración {i+1}")
    print(f"w = {w:.4f}, b = {b:.4f}")
    print(f"MSE = {mse:.4f}")
    print("-" * 30)
```

```python
print(f"Parámetros finales: w = {w:.4f}, b = {b:.4f}")
```

```python
# Ahora se puede predecir con los parámetros finales
y_pred_final = w * 9 + b
y_pred_final
```
